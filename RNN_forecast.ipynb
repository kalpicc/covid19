```python
filtered_df = df[df['unvaccinated_cases_per_100k'].notnull()]

filtered_df
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
data_unvaccinated = filtered_df[['unvaccinated_cases_per_100k']].values

data_unvaccinated
```

```python
data_vaccinated = filtered_df[['vaccinated_cases_per_100k']].values

data_vaccinated
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_vaccinated = scaler.fit_transform(data_vaccinated)
scaled_unvaccinated = scaler.fit_transform(data_unvaccinated)
```

```python
def sequence_data(data, time_step):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):  #for (i = 0; i <= 89 ;i++) 
        X.append(data[i:(i + time_step), 0])
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

```

```python
split_index = int(len(scaled_unvaccinated)*0.8)
split_i = int(len(scaled_vaccinated)*0.8)

train_unvaccinated = scaled_unvaccinated[:split_index]
test_unvaccinated = scaled_unvaccinated[split_index:]

train_vaccinated = scaled_vaccinated[:split_index]
test_vaccinated = scaled_vaccinated[split_index:]
```

```python
X_train, y_train = sequence_data(train_unvaccinated, 10)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

X_test, y_test = sequence_data(test_unvaccinated, 10)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
X_train1, y_train1 = sequence_data(train_vaccinated, 10)
X_train1 = X_train1.reshape(X_train1.shape[0], X_train1.shape[1], 1)

X_test1, y_test1 = sequence_data(test_vaccinated, 10)
X_test1 = X_test1.reshape(X_test1.shape[0], X_test1.shape[1], 1)
```

```python
X_train.shape
```

```python
X_train1.shape
```

```python
model = Sequential()
model.add(LSTM(50, input_shape=(10, 1)))
model.add(Dense(1))


# optimizer = Adam(learning_rate = 0.001)
from tensorflow.keras.optimizers import Adam

model.compile(optimizer= Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
model1 = Sequential()
model1.add(LSTM(50, input_shape=(10, 1)))
model1.add(Dense(1))


# optimizer = Adam(learning_rate = 0.001)
from tensorflow.keras.optimizers import Adam

model1.compile(optimizer= Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

history1 = model1.fit(X_train1, y_train1, epochs=100, batch_size=32, validation_data=(X_test1, y_test1))
```

```python
predictions = model.predict(X_test)

plt.figure(figsize=(14, 7))
plt.plot(y_test, label='Actual Values', color='blue')
plt.plot(predictions, label='Predicted Values', color='red')
plt.title('Time Series Forecast: Actual vs Predicted')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()
plt.show()
```
