```python
filtered_df = df[df['unvaccinated_cases_per_100k'].notnull()]

filtered_df
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
data = filtered_df[['unvaccinated_cases_per_100k']].values

data
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)

scaled_data
```

```python
def sequence_data(data, time_step):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):  #for (i = 0; i <= 89 ;i++) 
        X.append(data[i:(i + time_step), 0])
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

```

```python
split_index = int(len(scaled_data)*0.8)

train_data = scaled_data[:split_index]
test_data = scaled_data[split_index:]
```

```python
X_train, y_train = sequence_data(train_data, 10)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

X_test, y_test = sequence_data(test_data, 10)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
X_train.shape
```

```python
X_test
```

```python
model = Sequential()
model.add(LSTM(50, input_shape=(10, 1)))
model.add(Dense(1))


# optimizer = Adam(learning_rate = 0.001)
from tensorflow.keras.optimizers import Adam

model.compile(optimizer= Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)

plt.figure(figsize=(14, 7))
plt.plot(y_test, label='Actual Values', color='blue')
plt.plot(predictions, label='Predicted Values', color='red')
plt.title('Time Series Forecast: Actual vs Predicted')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()
plt.show()
```
